```python
import cv2 
from google.colab.patches import cv2_imshow
import numpy as np 

image = cv2.imread('/content/teethIdentify1.png') 
cv2_imshow(image)
```

```python
lower_green = np.array([0, 0, 0], dtype = "uint8") 

upper_green= np.array([0,255,0], dtype = "uint8")
```

```python
mask = cv2.inRange(image, lower_green, upper_green)
```

```python
detected_output = cv2.bitwise_and(image, image, mask =  mask) 

cv2_imshow(detected_output) 

cv2.waitKey(0) 
```
